# Key Findings and Insights

## 1. Overview

This notebook summarizes the most important discoveries from the West Midlands crime analysis, covering the period June 2024 – July 2025.  
The dataset contained approximately 385,000 records of reported crimes, including details on crime type, location, outcome, and temporal patterns.

The purpose of this summary is to highlight the key insights, statistical evidence, and visual confirmation of trends observed throughout the analysis.

---

## 2. Crime Trends Over Time

**Monthly crime counts show a clear seasonal pattern.**

- **Summer months** experience peaks in reported crime, with **July 2024 recording the highest volume (~31k crimes)**.
- **Winter months**, particularly **February 2025**, recorded the lowest volume (~24k crimes).
- The seasonal trend was tested statistically using ANOVA, confirming that seasonal differences are significant (**p < 0.05**). This means fluctuations are unlikely due to chance, they reflect real patterns in how crime varies across the year.

**Interpretation:**  
These findings are consistent with criminological research, which often observes that warmer weather contributes to increased public activity and, consequently, higher crime rates.  
The winter drop in crime may reflect fewer people outdoors and greater deterrence from poor weather conditions.

**Visuals referenced:**  
- **Line chart:** Monthly crime counts (clearly annotated peaks and troughs)  
- **Boxplot:** Seasonal distribution of crime (color-coded by season, with median and outlier markers)

## 3. Crime Types

**The most common categories of crime were:**

- **Violence and sexual offences** (~155k incidents)
- **Shoplifting** (~39k incidents)
- **Vehicle crime** (~33k incidents)
- Other categories such as burglary, public order offences, and drug-related crimes appeared at lower levels but still represent notable volumes.

A **Spearman correlation analysis** showed that:
- **Drug-related crimes** exhibited a strong positive trend over time (**ρ ≈ 0.73**).
- Some crime types (e.g., burglary, vehicle crime) showed negative correlations, suggesting a decline over the period.

**Interpretation:**  
Violence and sexual offences continue to dominate police records, indicating a persistent challenge for law enforcement and community safety.  
The upward trend in drug-related crimes highlights potential shifts in substance-related activity, which may need targeted policing or public health interventions.

**Visuals referenced:**  
- **Horizontal bar chart:** Top 10 crime types (with color-coded bars and value labels)
- **Line chart:** Trends for selected crime types (overlayed for comparison, with trend lines and confidence intervals)

## 4. Geographic Hotspots

**Analysis of LSOA-level data revealed that crime is highly concentrated in urban areas.**

- **Birmingham (LSOA 138A)** stood out as the single most concentrated hotspot (~6,800 crimes).
- Other hotspots were distributed across **Walsall, Coventry, Solihull, and Wolverhampton**, reinforcing the idea that urban density correlates with higher crime volumes.

**Location categories showed that:**
- **Supermarkets, shopping areas, petrol stations, and parking areas** are key hotspots for crime.
- These environments attract large public footfall, increasing opportunities for theft and anti-social behaviour.

**Interpretation:**  
The strong geographic clustering suggests targeted interventions in specific communities and location types could yield disproportionate crime reduction benefits.

**Visuals referenced:**  
- **Bar chart:** Top 10 LSOAs by crime count (with city labels and color gradient)
- **Interactive map:** Crime density heatmap (zoomable, with tooltips for LSOA and count)
- **Stacked bar chart:** Crime by location type (highlighting supermarkets, parking areas, etc.)

## 5. Outcomes of Crimes

**Analysis of case outcomes revealed a major systemic issue:**

- Only **~3.4% of crimes were solved**, meaning they led to a resolution such as caution, charge, or court result.
- In contrast, the vast majority remain unresolved:
    - **~171k cases ended with "Unable to prosecute suspect"**
    - **~111k cases with "No suspect identified"**

A **Chi-square test** confirmed that outcomes differ significantly by crime type (**p < 0.05**), meaning some crimes are statistically more likely to remain unsolved.

**Interpretation:**  
This raises concerns about investigative capacity and resource allocation, as a significant portion of crimes either fail to identify suspects or collapse at prosecution stage.  
Violent offences, despite their severity, often fall into unresolved categories, highlighting potential gaps in the justice pipeline.

**Visuals referenced:**  
- **Donut chart:** Solved vs unsolved cases (with percentage labels)
- **Heatmap:** Resolution rates by crime type (color-coded, with clear axis labels and legend)

---

## 6. Location-Specific Insights

A **z-test for proportions** confirmed that **shoplifting is disproportionately concentrated in supermarkets** (**p < 0.05**).

- Although shoplifting occurs in multiple locations, the supermarket environment accounts for a significantly higher share than would be expected by chance.
- **Parking areas and supermarkets** ranked as the top two most crime-prone environments overall, consistent with their accessibility and frequent public use.

**Interpretation:**  
These insights can directly inform place-based policing or environmental design interventions (e.g., CCTV in supermarkets, improved lighting in parking areas).

**Visuals referenced:**  
- **Grouped bar chart:** Shoplifting by location type (with supermarkets highlighted)
- **Bar chart:** Top 10 locations for all crimes (ranked and color-coded)

---

## 7. Overall Insights

- **Seasonality** is a major driver of crime, with higher volumes in summer.
- **Crime type composition** shows dominance of violent offences and notable growth in drug-related crimes.
- **Geography matters:** crimes cluster disproportionately in certain LSOAs and location types.
- **Justice outcomes are poor**, with over 96% of crimes unsolved.
- **Specific risks** include shoplifting in supermarkets and thefts in parking areas.

**Broader Implications:**

- Seasonal peaks may justify seasonal resourcing strategies (e.g., more patrols in summer).
- Rising drug-related crimes suggest a need for integrated policing and health approaches.
- The justice system’s resolution gap may undermine public trust and deterrence, requiring policy-level interventions.
- Place-specific strategies (e.g., hotspot policing, environmental crime prevention) are likely to be more effective than uniform approaches.
